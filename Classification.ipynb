```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from sklearn.metrics import classification_report
from tensorflow.keras.utils import to_categorical
import numpy as np


 #loading datasets
data=pd.read_csv('heart_failure.csv')

 #information about the data
print(data.info())

#3
from collections import Counter
print('Classes and number of values in the dataset',Counter(data['death_event']))

#4
#Extract the label column death_event and assign it to y
y = data['DEATH_EVENT']

#5
#Extract the features columns assign to x

x = data[['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']]

#6 get dummies
x = pd.get_dummies(x)


#7 splitting data
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)


#8 Initialize a ColumnTransformer object
ct = ColumnTransformer([('numeric',StandardScaler(),['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium','time'])])

#9
X_train = ct.fit_transform(X_train)

#10
X_test = ct.transform(X_test)

#11
le = LabelEncoder()

#12
Y_train = le.fit_transform(Y_train.astype(str))

#13
Y_test = le.fit_transform(Y_test.astype(str)) 

#14
Y_train = to_categorical(Y_train)

#15
Y_test = to_categorical(Y_test)


# 16 Initialize a tensorflow.keras.models.Sequential model instance

from tensorflow.keras.models import Sequential
model= Sequential()

#17
model.add(InputLayer(input_shape=(X_train.shape[1],)))

#18
model.add(Dense(12,activation='relu'))

#19
model.add(Dense(2, activation='softmax'))

#20
model.compile(loss='categorical_crossentropy',optimizer='adam', metrics=['accuracy'])

# 21  Using the Model.fit() function, fit the model instance model to the training data X_train and training labels Y_train

model.fit(X_train,Y_train,epochs = 100, batch_size=6, verbose=1)

#22 Using the Model.evaluate() function, evaluate the trained model instance model on the test data X_test and test labels Y_test

loss,acc = model.evaluate(X_test,Y_test,verbose=0)
print("Loss", loss, "Accuracy:", acc)

#23 Use the Model.predict() to get the predictions for the test data X_test with the trained model instance model

#24
y_estimate = model.predict(X_test,verbose=0)
y_estimate = np.argmax(y_estimate,axis=1)
#25
y_true = np.argmax(Y_test, axis=1)


#26
f1_score = classification_report(y_true,y_estimate)
print(f1_score)
```
